### Using numpy where

using list comprehension, we can determine what output, but the numpy where method can also do something very similar

```python
A = [1,2,3,4]
B = np.multiply(A, 4)
condition = [True, True, False, False]
list_cond = [(A_val if cond else B_val) for A_val,B_val, cond in zip(A,B, condition)]
list_cond
```

the same thing using numpy.where(condition, trueval, falseval) 

```python
np.where(condition, A, B)
```

you can also select using numpy.where

```python
np.where(B < 10, -1, B)
```
